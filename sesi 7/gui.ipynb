```python
# Mengenal Widget Dasar GUI (Label, Tombol, Input Teks)
import tkinter as tk
# buat window
window = tk.Tk()
window.title("Widget Dasar")
window.geometry("300x200")
# widget label
label = tk.Label(window, text="Halo, ini Label!")
# layout
label.pack(pady=10)

# widget entry (input teks)
entry = tk.Entry(window, width=30)
entry.pack(pady=10)

# button widget (tombol)
button = tk.Button(window, text="Tombol Button")
button.pack(pady=10)

# window main loop
window.mainloop()
```

```python
# Contoh grid()
import tkinter as tk
window = tk.Tk()
window.title("Layout Grid")
window.geometry("300x200")
# label dan entry dalam grid
tk.Label(window, text="Nama: ").grid(
    row=0,
    column=0,
    padx=5,
    pady=5
)
tk.Entry(window).grid(
    row=0, column=1,padx=5, pady=5
)
tk.Label(window, text="Umur: ").grid(
    row=1, column=0, padx=5, pady=5
)
tk.Entry(window).grid(
    row=1, column=1,padx=5,pady=5
)
tk.Button(window, text="Simpan").grid(row=2, column=0,
            columnspan=2, pady=10)
# tombol menempati 2 kolom

window.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox # untuk pop up error

# function hitung pembagian
def hitung():
    try:
        a = float(entry1.get())
        b = float(entry2.get())
        hasil = a / b
        label_hasil.config(text=f"Hasil: {hasil}") # configurasi/
    except ValueError:
        messagebox.showerror("Input Salah!", "Harap Masukkan Angka yg valid!")

    except ZeroDivisionError:
        messagebox.showerror("error matematika", "Tidak Bisa membagi dengannol")

window = tk.Tk()
window.title("error handling demo")
window.geometry("300x150")

# buatkan input angka
entry1 = tk.Entry(window, width=20)
entry1.pack(pady=5)
entry2 = tk.Entry(window, width=20)
entry2.pack(pady=5)

# tombol hitung
button = tk.Button(window, text='Hitung', command=hitung)
button.pack(pady=5)
# label hasil
label_hasil = tk.Label(window, text="Masukkan dua angka lalu klik hitung")
label_hasil.pack(pady=5)

window.mainloop()
```

```python
from tkinter import Tk
from tkhtmlview import HTMLLabel

root = Tk()
root.title("HTML in Tkinter")

html_content = """
<html>
<head>
<style>
body { font-family: Arial, sans-serif; color: #333; }
h1 { color: #007bff; }
p { line-height: 1.5; }
</style>
</head>
<body>
<h1>Welcome to my Tkinter App!</h1>
<p>This is some <b>HTML</b> content rendered within a <i>Tkinter</i> window.</p>
</body>
</html>
"""

html_label = HTMLLabel(root, html=html_content)
html_label.pack(fill="both", expand=True)

root.mainloop()
```
